```python
import pandas as pd

df = pd.read_csv("Placement_Data_Full_Class.csv")
```

```python
df.head()
```

```python
# Display basic information about the dataset
df.info()
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Check for duplicate records
df.duplicated().sum()
```

```python
# Fill missing salary values with 0
# Students who were not placed do not have a salary.

df['salary'] = df['salary'].fillna(0)
```

```python
# Check missing values again

df.isnull().sum()
```

```python
# Clean text columns

text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    df[col] = df[col].str.strip().str.title()
```

```python
# Rename columns for better readability

df.rename(columns={
    'ssc_p': 'ssc_percentage',
    'hsc_p': 'hsc_percentage',
    'degree_p': 'degree_percentage',
    'etest_p': 'etest_percentage',
    'mba_p': 'mba_percentage'
}, inplace=True)
```

```python
# Create a new column based on salary

df['salary_category'] = df['salary'].apply(
    lambda x: 'High' if x >= 500000 else ('Medium' if x > 0 else 'Not Placed')
)
```

```python
# Save the cleaned dataset

df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset saved successfully!")
```
